```python
class Node:
    def __init__(self,val=0,next=None):
        self.val=val
        self.next=next
class cll:
    def __init__(self,last=None):
        self.last=last
    def print(self):
        if self.last is None:
            print("List is empty")
            return
        
        temp = self.last.next   # first node
        start = temp

        while True:
            print(temp.val, end=" -> ")
            temp = temp.next
            if temp == start:
                break
        
        print("(circular)")
l=cll()
l.last=Node(1)
l.last.next=Node(2)
l.last.next.next=Node(3)
l.last.next.next.next=l.last

l.print()
```

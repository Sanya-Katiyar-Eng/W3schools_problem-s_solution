```python
# Using Singly linked list
class Node:
    def __init__(self,val=None,next=None):
        self.val=val
        self.next=next
class stack:
    def __init__(self,start=None):
        self.start=start
    def is_empty(self):
        if self.start==None:
            return True
        return False
    def push(self,val):
        n=Node(val,self.start)
        self.start=n
    def peak(self):
        if self.is_empty():
            return "stack is empty"
        return self.start.val
    def pop(self):
        if self.is_empty():
            return "stack is empty"
        self.start=self.start.next
    def size(self):
        count=0
        while self.start:
            count=count+1
            self.start=self.start.next
        return count
        

l=stack()
l.push(4)
l.push(12)
l.push(1)
print(l.peak())
l.pop()
print(l.peak())
l.size()
```

```python
#Reverse a string using stack
class Stack:
    def reverse_str(self,s):
        stack=[]
        rs=""
        for i in s:
            stack.append(i)
        for j in range(len(stack)):
            rs=rs+stack[-1]
            stack.pop()  
        return rs
        
l1=Stack()
name="sanya"
l1.reverse_str(name)
```

```python
#Next Greater Element
def greater_ele(s):
    stack=[]
    for i in range(len(s)):
        found=False
        j=i+1
        for j in range(j,len(s)):
            if s[j]>s[i]:
                stack.append(s[j])
                found=True
                break
        if not found:
            stack.append(-1)
    return stack
list=[4,5,2,10]
greater_ele(list)
```

```python
# previus greter element
def previous_ele(s):
    r=[]
    for i in range(len(s)):
        found=False
        for j in range(i-1, -1, -1):
            if s[i]<s[j]:
                r.append(s[j])
                found=True
                break
        if not found:
            r.append(-1)
    return r
                

list=[4,5,2,10]
previous_ele(list)
```

```python
# Largest Rectangle in Histogram
def rectangle(arr):
    s=[]
    for i in range(len(arr)):
        s.append(arr[i])
        if arr[i]<arr[i+1]:
            i=i+1
            s.append(arr[i])
        else :
            pass
        if arr[i]<arr[i-1]:
            i=i-1
            s.append(arr[i])
            i=i+1
    return s
        
        
a=[2,1,5,6,2,3]
rectangle(a)
```

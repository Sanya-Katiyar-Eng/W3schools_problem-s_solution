```python
# Using Singly linked list
class Node:
    def __init__(self,val=None,next=None):
        self.val=val
        self.next=next
class stack:
    def __init__(self,start=None):
        self.start=start
    def is_empty(self):
        if self.start==None:
            return True
        return False
    def push(self,val):
        n=Node(val,self.start)
        self.start=n
    def peak(self):
        if self.is_empty():
            return "stack is empty"
        return self.start.val
    def pop(self):
        if self.is_empty():
            return "stack is empty"
        self.start=self.start.next
    def size(self):
        count=0
        while self.start:
            count=count+1
            self.start=self.start.next
        return count
        

l=stack()
l.push(4)
l.push(12)
l.push(1)
print(l.peak())
l.pop()
print(l.peak())
l.size()
```

Singly Linked List Creation

```python
class Node:
    def __init__(self,data=0,next=None):
        self.data=data
        self.next=next
class sll:
    def __init__(self,start=None):
        self.start=start
    def show(self):
        temp=self.start
        while temp is not None:
            print(temp.data,end='->')
            temp=temp.next
        print("None")
l1=sll()
l1.start=Node(5)
l1.start.next=Node(2)
l1.start.next.next=Node(1)
l1.show()
```

Singly Linked List Size

```python
class Node:
    def __init__(self,data=0,next=None):
        self.data=data
        self.next=next
class SLL:
    def __init__(self,start=None):
        self.start=start
    def len1(self):
        count=0
        temp=self.start
        while temp is not None:
            count=count+1
            temp=temp.next
        return count
    
l=SLL()
l.start=Node(1)
l.start.next=Node(2)
l.start.next.next=Node(2)
l.start.next.next.next=Node(2)
l.start.next.next.next.next=Node(2)
l.start.next.next.next.next.next=Node(2)
print(l.len1())
```

Search in Singly Linked List

```python
class Node:
    def __init__(self,data=0,next=None):
        self.data=data
        self.next=next
class sll:
    def __init__(self,start=None):
        self.start=start
    def search(self,ele):
        temp=self.start
        while temp is not None:
            if temp.data==ele:
                return True
            temp=temp.next
        return False
l=sll()
l.start=Node(10)
l.start.next=Node(21)
l.start.next.next=Node(11)
l.start.next.next.next=Node(7)
l.start.next.next.next.next=Node(9)
print(l.search(10))
print(l.search(5))
```

```python
#Detect loop in linked list
class Node:
    def __init__(self,data=0,next=None):
        self.data=data
        self.next=next
class sll:
    def __init__(self,head=None):
        self.head=head
    def detect_loop(self):
        slow=self.head
        fast=self.head
        while fast is not None and fast.next is not None:
            slow=slow.next
            fast=fast.next.next
            if slow==fast:
                return True
        return False
   #.................................................... 
    def cycle_len(self):
        slow=self.head
        fast=self.head
        count=0
        while fast is not None and fast.next is not None:
            slow=slow.next
            fast=fast.next.next
            if slow==fast:
                count=count+1
            else:
                count=count-1
                fast=fast.next
                slow=slow.next

        return count

l=sll() 
l.head = Node(1)
l.head.next = Node(2)
l.head.next.next= Node(3)
l.head.next.next.next = Node(4)
l.head.next.next.next.next = Node(5)
l.head.next.next.next.next.next=l.head.next.next

l.detect_loop()
l.cycle_len()
```

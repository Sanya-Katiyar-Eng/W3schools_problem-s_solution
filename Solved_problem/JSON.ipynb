 Convert JSON Data to Python Object

```python

import json
json_obj =  '{ "Name":"David", "Class":"I", "Age":6 }'
print("Json dict :",json_obj)
python_obj = json.loads(json_obj)
print("python obj:",python_obj)
```

Convert Python Object to JSON Data

```python
import json 
python_dict={"name":"sanya","age":"18","marks":"75"}
print("python_dict :",python_dict,"type :",type(python_dict))
joso=json.dumps(python_dict)
print("python_dict :",joso,"type :",type(joso))
```

Convert Python Objects into JSON Strings

```python
import json 
python_dict={"name":"sanya","age":"18","marks":"75"}
print("python_dict :",python_dict,"\ntype :",type(python_dict))

joso=json.dumps(python_dict)
print("josn  :",joso,"\ntype :",type(joso))
python_dict =  {"name": "David", "age": 6, "class":"I"}
python_list =  ["Red", "Green", "Black"]
python_str =  "Python Json"
python_int =  (1234)
python_float =  (21.34)
python_T =  (True)
python_F =  (False)
python_N =  (None)

json_dict = json.dumps(python_dict)
json_list = json.dumps(python_list)
json_str = json.dumps(python_str)
json_num1 = json.dumps(python_int)
json_num2 = json.dumps(python_float)
json_t = json.dumps(python_T)
json_f = json.dumps(python_F)
json_n = json.dumps(python_N)

print("json dict : ", json_dict,"\ntype",type(json_dict))
print("jason list : ", json_list,"\ntype",type(json_list))
print("json string : ", json_str,"\ntype",type(json_str))
print("json number1 : ", json_num1,"\ntype",type(json_num1))
print("json number2 : ", json_num2,"\ntype",type(json_num2))
print("json true : ", json_t,"\ntype",type(json_t))
print("json false : ", json_f,"\ntype",type(json_f))
print("json null ; ", json_n,"\ntype",type(json_n))
```

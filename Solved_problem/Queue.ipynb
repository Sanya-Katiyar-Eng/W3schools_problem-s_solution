# Queue

```python
print("sanya")
#  Implement Queue using List
class Queue:
    def __init__(self):
        self.q=[]
    def is_empty(self):
        return len(self.q)==0
    def enqueue(self,val):
        self.q.append(val)
    def dequeue(self):
        if self.is_empty():
            raise IndexError("Queue Underflow")
        else:
            self.q.pop(0)
    def front(self):
        if self.is_empty():
            raise IndexError("Queue Underflow")
        else:
            return self.q[0]
    def rear(self):
        if self.is_empty():
            raise IndexError("Queue Underflow")
        else:
            return self.q[-1]
    def len(self):
        return len(self.q)
    def reverse(self):
        return self.q[::-1]
    
q=Queue()
q.enqueue(6)
q.enqueue(5)
q.enqueue(1)
q.enqueue(9)
print("Len : ",q.len())
print("front :",q.front())
print("rear :",q.rear())
q.dequeue()
print("front after dequeue :",q.front())
q.enqueue(0)
q.enqueue(6)
print(q.reverse())
```

Diamond Pattern

```python
for i in range(0,5):
    for j in range(0,i):
        print("*",end=" ")
    print("")
for i in range(-5,-1):
    for j in range (i,-1):
        print("*",end=" ")
    print("")
```

```python
number=(1,2,3,4,5,6,7,8,9)
counteven=0
countodd=0
for i in range(0,len(number)):
    if number[i]%2==0:
        counteven+=1
    else:
        countodd+=1
print("even num ",counteven,"\n odd num",countodd)
```

Count digits and letters in a string

```python
count_l,count_d,count_s=0,0,0
str=input("enter any mixed string : ")
for i in range(0,len(str)):
    if '9'>=str[i]>='0':
       count_d+=1
    elif 'a'<=str[i]>='z' or 'A'<=str[i]>='Z':
        count_l+=1
    else:
        count_s+=1
print("Original string : ",str)
print("Digits in string : ",count_d)
print("Letters in string : ",count_l)
print("Special word in string : ",count_s)

```

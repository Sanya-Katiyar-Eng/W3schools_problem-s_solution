Filter 4-Digit Binary numbers Divisible by 5

```python
n=int(input("How many num you want to take input ? Enter the num :"))
l1=[]
l2=[]
dec,j,k=0,0,0
for i in range(0,n):
    num1=int(input("Enter the binary num :"))
    l1.append(num1)
print(l1)
for num in l1:
    digits=num
    while num!=0:
        d=num%10
        num=int(num/10)
        dec+=(2**j)*d
        j=j+1
    l2.append(dec)
    if dec%5==0:
        print("divisible by 5 :",digits)
    dec=0
    j=0

```

Count digits and letters in a string

```python
count_l,count_d,count_s=0,0,0
str=input("enter any mixed string : ")
for i in range(0,len(str)):
    if '9'>=str[i]>='0':
       count_d+=1
    elif 'a'<=str[i]>='z' or 'A'<=str[i]>='Z':
        count_l+=1
    else:
        count_s+=1
print("Original string : ",str)
print("Digits in string : ",count_d)
print("Letters in string : ",count_l)
print("Special word in string : ",count_s)

```

## Dog age in dog years
  
Note: For the first two years, a dog year is equal to 10.5 human years. After that, each dog year equals 4 human years.


```python
#2*10.5+13*4
dog_age=int(input("Enter the dog's age :"))
print(" dog's age :",dog_age)
if dog_age>2:
    print("The dog's age in dog's year is :",21+(dog_age-2)*4)
elif dog_age==1:
    print("The dog's age in dog's year is :",10.5)
else:
    print("The dog's age in dog's year is :",21)
```

Uppercase string if 2+ uppercase chars in first 4.

```python
str=input("enter string : ")
count=0
print("Original string : ",str)
if len(str)>=4:
    for i in range(0,5):
        if 'Z'>=str[i]>='A':
            count=count+1
            if count>=2:
                print("Uppercase string :",str.upper())
                break
else:
    for i in range(0,len(str)):
        if 'Z'>=str[i]>='A':
            count=count+1
            if count>=2:
                print("Uppercase string :",str.upper())
                break
```

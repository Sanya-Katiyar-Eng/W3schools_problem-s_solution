Sum of list

```python
def list_sum(list):
    count=0
    for i in range(0,len(list)):
        count+=list[i]
    print("Sum of list :",count)
list=[2,4,5,6,7]
list_sum(list)
```

Fabonacci Sequence

```python
def fab(num):
    if num==0:
        return 0
    elif num==1 or num==2:
        return 1
    else:
        return  fab(num-1)+fab(num-2)
num=int(input("enter the num :"))
fab(num)
```

Factorial Using Recursion

```python
def fact(num):
    if num==0:
        return 1
    else:
        return fact(num-1)*num
num=int(input("enter the num :"))
print("Factorial of ",num,"is :",end=" ")
fact(num)
```

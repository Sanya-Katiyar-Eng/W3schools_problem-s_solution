Sum of list

```python
def list_sum(list):
    count=0
    for i in range(0,len(list)):
        count+=list[i]
    print("Sum of list :",count)
list=[2,4,5,6,7]
list_sum(list)
```

Fabonacci Sequence

```python
def fab(num):
    if num==0:
        return 0
    elif num==1 or num==2:
        return 1
    else:
        return  fab(num-1)+fab(num-2)
num=int(input("enter the num :"))
fab(num)
```

Factorial Using Recursion

```python
def fact(num):
    if num==0:
        return 1
    else:
        return fact(num-1)*num
num=int(input("enter the num :"))
print("Factorial of ",num,"is :",end=" ")
fact(num)
```

Sum of Nested Lists Using Recursion

```python
def sum_lists(list):
    count=0
    for element in list:
        if type(element)==type([]):
            count=count+sum_lists(element)
        else:
            count=count+element
    return count
L=[1,2,[3,4],5,[6,7]]
sum_lists(L)
```

 Sum of Digits of an Integer Using Recursion

```python
def Sum_digit(d):
    if d==0:
        return 0
    else:
        return Sum_digit(int(d/10))+(d%10)
    

digit=int(input("Enter the digit :"))
print(Sum_digit(digit))
```

Sum of Series n + (n-2) + (n-4) ... Using Recursion

```python
def sum_series(n):
    print(n)
    if n==0:
        return 0
    else:
        return n+sum_series(int(n-2))
num=int(input("Enter the num :"))
print("sum is : ",sum_series(num))
```

 Get Largest Number in List

```python
l=[23,4,5,89,1]
print("Max num :",max(l))
```

```python
max=0
for i in range(len(l)):
    if l[i]>max:
        max=l[i]
print("max without using inbuild function :",max)
```

Get Smallest Number in List

```python
print("Min num :",min(l))
```

```python
min=l[0]
for i in range(len(l)):
    if l[i]<min:
        min=l[i]
print("Without using inbuild function :",min)
```

Count Strings with Same Start and End

```python
List = ['abc', 'xyz', 'aba', '1221']
count=0
for i in range(len(List)):
    a=List[i]
    if a[0]==a[len(a)-1]:
            count+=1

print(count)
```

Pandas problem

Creating a Series

```python
import pandas as pd
a=(1,2,3,4,5)#tuple
print("using tuple:\n",pd.Series(a))
b=[9,6,3,2]#list
print("using list\n",pd.Series(b))
c={1:"sanya",2:"Anmol"}
print("Using dict  :\n",pd.Series(c))

```

Series to List Conversion

```python
a=[1,2,3,4]
s=pd.Series(a)
b=list(a)
print(b)
print(type(b))
```

Series Arithmetic

```python
a=pd.Series([2, 4, 6, 8, 10])
b=pd.Series([1, 3, 5, 7, 9])
print("Add :\n",a+b)
print("Subtract :\n",a-b)
print("Multiple :\n",a*b)
print("Divide :\n",a/b)


```

Series Comparison

```python
a=pd.Series([2, 4, 6, 8, 10])
b=pd.Series([1, 3, 5, 7, 10])
print("Equal:",a==b)
print("greter then :\n",a>b)
print("less then :\n",a<b)
print("greter then equal :\n",a>=b)
print("less then equal :\n",a<=b)
```

Dict to Series

```python
dict={'a': 100, 'b': 200, 'c': 300, 'd': 400, 'e': 800}
se=pd.Series(dict)
se
```

 Array to Series

```python
arr=[10,20,30, 40,50]
print(pd.Series(arr))
```

Change Series DataType

```python
s=pd.Series([100,200,"Python",300.12,400])
print("Original Series :")
print(s)
print("Change the said data type to numeric")
c=pd.to_numeric(s,errors='coerce')
print(c)
```

DF Column to Series

```python
data=pd.DataFrame({'col1':[1,2,3,4,7,11],'col2':[4,5,6,9,5,0],'col3':[7,5,8,12,1,11]})
print("Datafarme :\n",data)
print("1st column as series :\n")
print(pd.Series(data['col1']))
```

Series to Array



```python
a=pd.Series([1,2,3,4])
print("Original series : \n")
print(a)
print("Series to Array \n")
arr=a.values
print(arr)
```

Flatten Series of Lists

```python
A=pd.Series({0:['Red','Green','White'],1:['Red','Green'],2:['Yellow']})
print("Original Series :\n",A)
print("One Series ")
b=A.values
print(b)
```

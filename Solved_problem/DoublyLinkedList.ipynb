```python
#simple printing Doubly linked list
class Node:
    def __init__(self,pre=None,val=0,next=None):
        self.pre=pre
        self.val=val
        self.next=next
class dll:
    def __init__(self,head=None):
        self.head=head
    def printForword(self):
        temp=self.head
        while temp is not None:
            print(temp.val,end="<->")
            temp=temp.next
        print("None")

    def printBackword(self):
        temp=self.head
        while temp.next is not None:
            temp=temp.next
        while temp is not None:
            print(temp.val,end="<->")
            temp=temp.pre
        print("None")

l=dll()
l.head=Node(val=1)

l.head.next=Node(val=3)
l.head.next.pre=l.head

l.head.next.next=Node(val=4)
l.head.next.next.pre=l.head.next
print("Print Forword : ",end=" ")
l.printForword()
print("Print Backward : ",end=" ")
l.printBackword()
```

```python
#Insert and delete..................................
class Node:
    def __init__(self,pre=None,val=None,next=None):
        self.pre=pre
        self.val=val
        self.next=next
class DLL:
    def __init__(self,head=None):
        self.head=head
    def printForward(self):
        temp=self.head
        while temp is not None:
            print(temp.val,end="<->")
            temp=temp.next
        print("None")
    def printBackword(self):
        temp=self.head
        while temp.next is not None:
            temp=temp.next
        while temp is not None:
            print(temp.val,end="<->")
            temp=temp.pre
        print("None")

    def insert_at_beginning(self, val):
        if self.head is None:
            self.head = Node(None, val, None)
        else:
            n = Node(None, val, self.head)  # new node
            self.head.pre = n               # old head ka pre update
            self.head = n                  # head shift
    def insert_at_last(self,val):
        temp=self.head
        while temp.next is not None:
            temp=temp.next
        n=Node(temp,val,None)
        temp.next=n
    def insert_at_position(self,val,pos):
        count=1
        temp=self.head
        while temp is not None:
            if count==pos:
                n=Node(temp.pre,val,temp)
                if temp.pre is not None:
                    temp.pre.next=n
                temp.pre=n
            temp=temp.next
            count=count+1
        return "pos is not found"
    def delete_at_beginning(self):
        self.head=self.head.next
        self.head.pre=None
    def delete_at_last(self):
        temp=self.head
        while temp.next is not None:
            temp=temp.next

        temp.pre.next=None
    def delete_at_position(self,pos):
        count=1
        temp=self.head
        while temp is not None and temp.next is not None:
            if count==pos:#1234
                temp.next.pre=temp.pre
                temp.pre.next=temp.next
            temp=temp.next
            count=count+1
    def count(self):
        count=0
        temp=self.head
        while temp is not None:
            temp=temp.next
            count=count+1
        return count
    def search(self,val):
        temp=self.head
        while temp is not None:
            if temp.val==val:
                return True
            temp=temp.next
        return False
    def mid(self):
        count=0
        c=self.count()
        m=c//2
        temp=self.head
        while temp is not None:
            if count==m:
                return temp.val
            temp=temp.next
            count=count+1
        return False



l=DLL()
l.head=Node(val=1)

l.head.next=Node(val=3)
l.head.next.pre=l.head

l.head.next.next=Node(val=1)
l.head.next.next.pre=l.head.next
l.printForward()
print("\n\nInsert : ")
print("At beginning :")
l.insert_at_beginning(2)
l.insert_at_beginning(9)
l.printForward()
print("At last :")
l.insert_at_last(2)
l.insert_at_last(11)
l.printForward()
l.printBackword()
print("At Position :")
l.insert_at_position(1,3)
l.printForward()
print("\n\nDeletion : ")
print("At beginning :")
l.delete_at_beginning()
l.printForward()
#l.printBackword()
print("At last :")
l.delete_at_last()
l.printForward()
#l.printBackword()
print("At Position :")
l.delete_at_position(2)
l.printForward()
#l.delete_at_position(3)
l.printForward()
print("\n\nNum of Nodes if DLL : ")
print(l.count())

print("\n Searching val")
l.search(5)
print("\n mid val")
l.mid()
```
